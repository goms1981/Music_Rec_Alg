## Univariate Exploratory Analysis (Single variable Exploration)
## 

### 1. Categorical Variables
### genre, topic,artist_name

```python
#Plot the number of songs in each genre 
genre_counts=Music['genre'].value_counts()

#Plot the barchart :
genre_counts.plot.bar()
plt.title('Genre Distribution')
plt.xlabel('Genre')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show
        

```

```python
#Plot number of songs in each topic
topic_counts=Music['topic'].value_counts()
#Plot the barh chart
topic_counts.plot(kind='barh', color='skyblue')
plt.title('Top 10 Topics')
plt.xlabel('Number of Songs')
plt.ylabel('Topic')
plt.tight_layout()
plt.show()
```

```python
#Plot Top 20 Artist by Number of songs
artist_counts=Music['artist_name'].value_counts().head(20)
artist_counts.plot(kind='barh', color='teal')
#Plot barh chart
plt.title('Top 20 Artist name ')
plt.xlabel('Number of Songs')
plt.ylabel('Artist Name')
plt.tight_layout()
plt.show()
```

```python
#Plot Top 20 Artist by Number of songs in pie chart

#Plot the piechart :
plt.figure(figsize=(6,6))
plt.pie(artist_counts, labels=artist_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Top 20 Artist by Number of songs')
plt.show()
```

## 2. Numerical Variable
### age,len,shake the audience,romantic,family/spiritual

```python
# Selecting numerical columns for distribution plots
numerical_cols = ['len','age','shake the audience','romantic','family/spiritual','world/life','movement/places','sadness','release_date',]

# Plotting distributions of numerical features
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 4, i + 1)
    sns.histplot(Music[col], kde=True)
    plt.title(col)
    plt.xscale('log')
    plt.tight_layout()
```

##### In the above chart "len"appears roughly normally distributed .
##### "age" is somewhat skewed .All the others are highly skewed ,most of the values are at very low end.

# Bivariate Analysis:

```python
plt.figure(figsize=(18, 12))
sns.boxplot(data=Music, x='genre', y='release_date')
plt.title('Distribution of Release Year by Genre')
plt.xlabel('Genre')
plt.ylabel('Release Year')
plt.xticks(rotation=45) 
```

###  Genre VS Release year: The Hip Hop genre has the highest median release year (around 2010-2011) and its entire Interquartile Range falls roughly between 2000 and 2017. This indicates that the majority of hip hop tracks in this dataset are from the 21st century.Overall, the plot shows clear differences in the typical release eras represented for each genre within this dataset, with a noticeable shift towards more recent years for genres like Reggae and especially Hip Hop compared to others like Country, Blues, or Jazz.

```python
# Compare romantic scores across genres
sns.boxplot(data=Music, x='genre', y='romantic')
plt.xticks(rotation=45)
plt.title("Romantic Score by Genre")
plt.show()
```

### While most songs aren't very romantic, some genres like jazz and country tend to be a little more romantic than others on average.

```python
# Relationship between sadness and romantic
sns.scatterplot(data=Music, x='romantic', y='sadness')
plt.title("Sadness vs. Romantic")
plt.show()
```

### Looks like more romantic songs tend to be less sad.

```python
# Correlation matrix of all numeric columns
numeric_cols = Music.select_dtypes(include='float64')
corr = numeric_cols.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Looks like songs high in feelings are often express Sadness--0.64.
### Energetic songs (shake the audience) tend to be about movement or places --0.60

```python
# Count of topics across genres (if there are few genres & topics)
pd.crosstab(Music['genre'], Music['topic']).plot(kind='bar', stacked=True)
plt.title("Genre vs Topic")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Genre Vs Topic: Pop has the highest overall count of topic instances. Genres like Blues, Country, Jazz, and Reggae show a more diverse mix of topics compared to the specialization seen in Hip Hop and Rock.Topics like 'sadness' and 'world/life' appear frequently across multiple genres.

```python
plt.figure(figsize=(12,6))
sns.countplot(data=Music, x='genre', hue='topic')
plt.xticks(rotation=45)
plt.title("Topic Counts by Genre")
plt.tight_layout()
plt.show()
```

### Pop and Country have the highest number of songs overall, especially on topics like sadness, world/life, and violence.Violence is the most common topic in rock music.Hip hop has a strong focus on obscene content compared to other genres.Romantic themes appear more in pop, country, and jazz.Feelings and night/time are less common across all genres.
